```python
#rebase function programming

dn = 42
l = []

for i in range(0, 1000*100) :
    v = dn // 2**i
    if v > 0:
        s = v % 2
        l.append(s)

sl = l[::-1]

ln = int("".join(map(str, sl)))
print(ln)
```

```python
def base_10_to_2(input_number):
  l = list()

  for i in range(0, 10**3):
    v = input_number // 2**i
    if v > 0:
      s = v % 2
      l.append(s)
      sl = l[::-1]
    ln = int("".join(map(str, sl)))
  return ln
```
